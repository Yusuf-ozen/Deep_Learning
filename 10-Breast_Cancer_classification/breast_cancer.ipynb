```python
data = data.drop("id", axis = 1)
data.head()
```

```python
y = data["diagnosis"]
x = data.drop("diagnosis", axis = 1)
```

```python
def function(x, y):
    lr = LogisticRegression()
    kn = KNeighborsClassifier()
    dt = DecisionTreeClassifier(criterion = "entropy")   #"gini", "entropy", "log_loss"
    br = BernoulliNB()
    gs = GaussianNB()
    sv = SVC()
    rf = RandomForestClassifier()
    
    models = [lr, kn, dt, br, gs, sv, rf]
    model_name = ["Logistic", "KNN", "Decision", "Bernoulli", "Gaussian", "SVM", "Random"]
    
    x_train, x_test, y_train, y_true = train_test_split(x, y, test_size = 0.2, random_state = 42)
    #x_train = MinMaxScaler().fit_transform(x_train)
    #x_test = MinMaxScaler().fit_transform(x_test)
    x_train = StandardScaler().fit_transform(x_train)
    x_test = StandardScaler().fit_transform(x_test)
    
    acc = []
    
    result = pd.DataFrame(columns = ["accuracy"], index = model_name)
    
    for i in models:
        y_pred = i.fit(x_train, y_train).predict(x_test)
        acc.append(accuracy_score(y_true, y_pred) * 100)
    
    result["accuracy"] = acc
    sira = result.sort_values("accuracy", ascending = False)
    
    return sira
```

```python
function(x, y)   ## KNN has best score = 95.61 in MinMaxScaler and logistic has best score in StandardScaler. Also Standard scaler 
# is better than MinMaxScaler() for this dataset
```
